# 코사인 유사도

````python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))

vec1 = np.array([0,1,1,1])
vec2 = np.array([1,0,1,1])
vec3 = np.array([2,0,2,2])

print('벡터1과 벡터2의 유사도 :',cos_sim(vec1, vec2))
print('벡터1과 벡터3의 유사도 :',cos_sim(vec1, vec3))
print('벡터2와 벡터3의 유사도 :',cos_sim(vec2, vec3))
````

# OpenAI의 Embedding API

````python
import os
import numpy as np
from numpy import dot
from numpy.linalg import norm
import pandas as pd
from langchain.embeddings import OpenAIEmbeddings

import os
from pathlib import Path
from dotenv import load_dotenv

def get_openai_key():

    # 현재 파일 위치
    current_path = Path.cwd()
    
    # step10_LLM 디렉토리로 이동 (../../)
    project_root = current_path.parent.parent
    
    # .env 파일 경로
    env_path = project_root / '.env'
    
    # .env 파일이 존재하는지 확인
    if env_path.exists():
        load_dotenv(env_path)
        print(f"Loaded .env from: {env_path}")
    else:
        raise FileNotFoundError(f".env file not found at: {env_path}")
    
    # 실행
    try:
        # 환경 변수 사용
        api_key = os.environ.get('OPENAI_API_KEY')
        if api_key:
            print("API Key loaded successfully")
            # API 키 마스킹하여 출력
            masked_key = f"{api_key[:8]}...{api_key[-4:]}"
            print(f"API Key: {masked_key}")
        else:
            print("API Key not found in .env file")
    except Exception as e:
        print(f"Error loading .env file: {e}")

    print("""Returned API Key: """, api_key)
    return api_key


os.environ['OPENAI_API_KEY'] = get_openai_key()
````

````python
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")
query_result = embeddings.embed_query('저는 배가 고파요')
print(query_result)
````

````python
data = ['저는 배가 고파요',
        '저기 배가 지나가네요',
        '굶어서 허기가 지네요',
        '허기 워기라는 게임이 있는데 즐거워',
        '스팀에서 재밌는 거 해야지',
        '스팀에어프라이어로 연어구이 해먹을거야']

df = pd.DataFrame(data, columns=['text'])

def get_embedding(text):
  return embeddings.embed_query(text)

df['embedding'] = df.apply(lambda row: get_embedding(
        row.text,
    ), axis=1)
df
````

````python
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))

def return_answer_candidate(df, query):
    query_embedding = get_embedding(
        query
    )
    df["similarity"] = df.embedding.apply(lambda x: cos_sim(np.array(x),
                                                            np.array(query_embedding)))
    top_three_doc = df.sort_values("similarity",
                                ascending=False).head(3)
    return top_three_doc

sim_result = return_answer_candidate(df, '아무것도 안 먹었더니 꼬르륵 소리가 나네')
sim_result
````

# ChatOpenAI와 Memory

````python
import os
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain.callbacks.tracers import ConsoleCallbackHandler

os.environ['OPENAI_API_KEY'] = get_openai_key()

# 객체 생성
llm = ChatOpenAI(
    temperature=0.1,  # 창의성 (0.0 ~ 2.0)
    max_tokens=2048,  # 최대 토큰수
    model_name="gpt-3.5-turbo",  # 모델명
)

# 질의내용
question = "세종대왕이 누구인지 설명해주세요"

# 질의
result = llm.invoke(question)
print(result.content)

# 질문 템플릿 형식 정의
template = "{who}가 누구인지 설명해주세요"

# 템플릿 완성
prompt = PromptTemplate(
        template=template, input_variables=['who']
    )
print(prompt)

print(prompt.format(who="오바마"))
````

````python
# 연결된 체인(Chain)객체 생성
llm_chain = prompt | llm
result = llm_chain.invoke({"who":"이순신 장군"})
print(result)
````

````python
result = llm_chain.invoke({"who":"이순신 장군"},
                          config={'callbacks': [ConsoleCallbackHandler()]})
````

````python
# 질문 템플릿 형식 정의
template= """당신은 위키북스에서 만든 친절한 위키봇입니다.

Current Conversation: {history}

Human: {input}

AI:"""

# 템플릿 완성
prompt = PromptTemplate(
        template=template, input_variables=['history', 'input']
    )
print(prompt)
````

````python
model = ChatOpenAI()
chain = prompt | model
````

````python
store = {}
````

````python
session_id = "test"
````

````python
# 세션 기록 가져오기
if session_id not in store:
    store[session_id] = ChatMessageHistory()

session_history = store[session_id]  # 안전하게 세션 기록을 변수에 할당

with_message_history = RunnableWithMessageHistory(
    chain,
    lambda session_id: session_history,  # 직접 참조를 반환하도록 수정
    input_messages_key="input",
    history_messages_key="history"
)
````

````python
# 주어진 메시지와 설정으로 체인을 실행합니다.
result = with_message_history.invoke(
    {"input": "당신은 어디에서 만들었습니까?"},
    config={"configurable": {"session_id": "test"}},
)
print(result.content)
````

# 의미로 분할하는 SemanticChunker

````python
import os
import urllib.request
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_experimental.text_splitter import SemanticChunker

urllib.request.urlretrieve("https://raw.githubusercontent.com/chatgpt-kr/openai-api-tutorial/main/ch06/test.txt", filename="test.txt")

with open("test.txt", encoding="utf-8") as f:
    file = f.read()
print('텍스트의 길이:', len(file))
````

- 문서 분할 방식
    - 백분위수(Percentile) 방식
    - 표준편차(Standard Deviation) 방식
    - 사분위수(Interquartile Range) 방식
- 위 3가지 방식 모두 코사인 거리라는 개념 사용 

## 백분위수 방식
- 모든 연속된 문장 쌍 사이의 코사인 거리를 계산하고 그 거리들의 95번째 백분위수를 임계값으로 설정
- 임계값보다 큰 코사인 거리를 가진 문장 쌍 사이에서 문서를 분할
- 즉, 의미적으로 가장 다른 문장들(상위 5%) 사이에서 분할이 일어남


````python
text_splitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_type="percentile",
    breakpoint_threshold_amount=95,
)
texts = text_splitter.create_documents([file])
print('분할된 청크의 수:', len(texts))
````

## 표준편차 방식
- 모든 연속된 문장 쌍 사이의 코사인 거리의 평균과 표준편차를 계산
- 평균으로부터 3 표준편차만큼 떨어진 거리를 임계값으로 설정
- 임계값보다 큰 코사인 거리를 가진 문장 쌍 사이에서 문서를 분할
- 즉, 의미적으로 매우 다른 문장들(통계적 이상치) 사이에서 분할이 일어남


````python
text_splitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_type="standard_deviation",
    breakpoint_threshold_amount=3,
)
texts = text_splitter.create_documents([file])
print('분할된 청크의 수:', len(texts))
````

## 사분위수 방식
- 모든 연속된 문장 쌍 사이의 코사인 거리를 계산
- 코사인 거리의 사분위수(Q1, Q3)와 사분위수 범위(IQR = Q3 - Q1)를 계산
- Q3 + 1.5 * IQR을 임계값으로 설정 (1.5는 조정 가능한 파라미터)
- 임계값보다 큰 코사인 거리를 가진 문장 쌍 사이에서 문서를 분할
- 즉, 의미적으로 매우 다른 문장들(통계적 이상치) 사이에서 분할이 일어남


````python
text_splitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_type="interquartile",
    breakpoint_threshold_amount=1.5
)
texts = text_splitter.create_documents([file])
print('분할된 청크의 수:', len(texts))
````

# 벡터 데이터베이스
- 벡터데이터베이스로는 Milvus, Chroma, Faiss, Pinecone 등이 있음 
- 본 장에서는 Chroma와 Faiss를 사용


````python
import os
import urllib.request
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.vectorstores import FAISS

loader = PyPDFLoader('2023_북한인권보고서.pdf')
pages = loader.load_and_split()
print('청크의 수:', len(pages))
````

````python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
splited_docs = text_splitter.split_documents(pages)
print('분할된 청크의 수:', len(splited_docs))
````

````python
chunks = [splited_doc.page_content for splited_doc in splited_docs]
print('청크의 최대 길이 :',max(len(chunk) for chunk in chunks))
````

````python
db = Chroma.from_documents(splited_docs, OpenAIEmbeddings())
print('문서의 수:', db._collection.count())
````

````python
question = '북한의 교육과정'
docs = db.similarity_search(question)
print('문서의 수:', len(docs))
````

````python
for doc in docs:
  print(doc)
  print('--' * 10)
````

````python
db_to_file = Chroma.from_documents(splited_docs, OpenAIEmbeddings(), persist_directory = './chroma_test.db')
print('문서의 수:', db_to_file._collection.count())
````

````python
db_from_file = Chroma(persist_directory='./chroma_test.db',
		      embedding_function=OpenAIEmbeddings())
````

````python
print('문서의 수:', db_from_file._collection.count())
````

````python
faiss_db = FAISS.from_documents(splited_docs, OpenAIEmbeddings())
print('문서의 수:', faiss_db.index.ntotal)
````

````python
faiss_db.save_local('faiss_index')

new_db_faiss = FAISS.load_local('faiss_index',
				OpenAIEmbeddings(),
				allow_dangerous_deserialization=True)
````

````python
question = '북한의 교육 과정'
docs = new_db_faiss.similarity_search(question)

for doc in docs:
  print(doc)
  print('--' * 10)
````
