# 마법공식

```python
import os 
import pandas as pd


xls = os.listdir("data/step06_v1_data")
dfs = []
for i in xls:
    df = pd.read_excel(f"data/step06_v1_data/{i}", index_col=0, dtype={"code": str})
    dfs.append(df)
df = pd.concat(dfs)

cond = df['day'] == "12월 결산"
df = df[cond]
```

```python
cond = (df['roic'] >0) & (df['ev/ebitda'] > 0)
```

```python
df2 = df[cond].copy()
df2
```

```python
df2['rank1'] = df2['roic'].rank(ascending=False)
df2
```

```python
df2['rank2'] = df2['ev/ebitda'].rank()
df2
```

```python
df2['rank'] = df2['rank1'] + df2['rank2']
df2
```
