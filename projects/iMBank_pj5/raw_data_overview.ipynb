```python
def analyze_changes(customer_id):
    # 해당 고객 데이터 가져오기
    customer_data = show_me(customer_id).sort_values('기준년월')
    
    # 수치형 컬럼들
    numeric_columns = ['수신_요구불예금', '수신_거치식예금', '수신_적립식예금', '수신_펀드', '수신_외화예금', '대출금액']
    
    analysis = {}
    
    for col in numeric_columns:
        data = customer_data[col]
        diff_data = data.diff()

        total_change = abs(diff_data).sum()
        direction_index = diff_data.sum() / total_change if total_change != 0 else 0
        
        changes = {
            '전체_변화량': data.iloc[-1] - data.iloc[0],  # 최종 - 최초
            '월평균_변화량': data.diff().mean(),  # 월별 변화의 평균
            '변화_횟수': (data.diff() != 0).sum(),  # 변화가 있었던 횟수
            '최대_상승폭': data.diff().max(),  # 가장 큰 증가
            '최대_하락폭': data.diff().min(),  # 가장 큰 감소
            '변동성': data.std(),  # 표준편차로 본 변동성
            '방향성_지수': direction_index,  # 변화의 방향성 (-1 ~ 1)
        }
        
        analysis[col] = changes
        analysis = pd.DataFrame(analysis)
    return analysis.fillna(0)

# 사용 예시
customer_changes = analyze_changes(4762826158)
customer_changes
```
