```python
def remove_outliers_all_columns(df, columns, k=1.5):
    """
    여러 컬럼의 이상치를 IQR 방법으로 제거하는 함수
    """
    df_clean = df.copy()
    
    for column in columns:
        Q1 = df_clean[column].quantile(0.01)
        Q3 = df_clean[column].quantile(0.8)
        IQR = Q3 - Q1
        lower_bound = Q1 - k * IQR
        upper_bound = Q3 + k * IQR
        
        # 이상치를 가진 행 제거
        df_clean = df_clean[(df_clean[column] >= lower_bound) & 
                           (df_clean[column] <= upper_bound)]
        
        print(f"\n{column} 컬럼 처리:")
        print(f"하한값: {lower_bound:.2f}")
        print(f"상한값: {upper_bound:.2f}")
        print(f"현재 데이터 개수: {len(df_clean)}")

    print(f"\n최종 결과:")
    print(f"원본 데이터 개수: {len(df)}")
    print(f"정제된 데이터 개수: {len(df_clean)}")
    print(f"제거된 행 개수: {len(df) - len(df_clean)}")
    
    return df_clean

# 이상치 제거 실행
columns_to_clean = ['age01','p02g17', 'p02g18', 'p02g19', 'p02g20', 
                   'p02g21', 'p02g22', 'p02g23', 'p02g25', 'p02g26']

# 이상치 제거 전 기초 통계량 확인
print("이상치 제거 전 기초 통계량:")
print(df[columns_to_clean].describe())

# 이상치 제거 실행
df_cleaned = remove_outliers_all_columns(df, columns_to_clean)

# 이상치 제거 후 기초 통계량 확인
print("\n이상치 제거 후 기초 통계량:")
print(df_cleaned[columns_to_clean].describe())

# 시각화를 통한 확인 (선택사항)
import matplotlib.pyplot as plt

def plot_boxplots(df_before, df_after, columns):
    n_cols = len(columns)
    fig, axes = plt.subplots(2, n_cols, figsize=(n_cols*4, 8))
    
    for i, col in enumerate(columns):
        # 이상치 제거 전
        axes[0,i].boxplot(df_before[col].dropna())
        axes[0,i].set_title(f'{col} (전)')
        
        # 이상치 제거 후
        axes[1,i].boxplot(df_after[col].dropna())
        axes[1,i].set_title(f'{col} (후)')
    
    plt.tight_layout()
    plt.show()

# 박스플롯 그리기
plot_boxplots(df, df_cleaned, columns_to_clean)
```
