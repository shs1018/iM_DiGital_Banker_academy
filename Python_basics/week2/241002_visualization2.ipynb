## Group by 연산
- 매우 중요하다!

```python
from pandas import DataFrame

data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
    ["해운", "팬오션", 21.23, 0.95],
    ["시스템반도체", "티엘아이", 35.97, 1.12],
    ["해운", "HMM", 21.52, 3.20],
    ["시스템반도체", "아이에이", 37.32, 3.55],
    ["2차전지(생산)", "LG화학", 83.06, 3.75]
]

columns = ["테마", "종목명", "PER", "PBR"]
df = DataFrame(data=data, columns=columns)
df
```

```python
res = df.groupby('테마')[['PER','PBR']].mean()
print(res)
```

```python
res2 = df.groupby("테마").get_group("시스템반도체")
print(res2)
```

- 칼럼추가 '분기','연도','월','일'

```python
set_lst = ['분기', '연도', '월', '일']

def set_time_cols(data):
  data['분기'] = data['일자'].dt.quarter
  data['연도'] = data['일자'].dt.year
  data['월'] = data['일자'].dt.month
  data['일'] = data['일자'].dt.day
  data = data.drop(columns = ['일자'])
  return data

df2 = set_time_cols(df)
df2.info()
```

```python
res = df2.groupby(['연도', '분기']).get_group((2021, 2)).mean()
print(res)
```

```python
multiples = {
    "시가":"first",
    "고가":max,
    "저가":min,
    "종가":'last',
    "거래량":sum
}

res = df2.groupby(['연도', '분기']).agg(multiples)
print(res.reset_index())
```

# 캔들차트

```python
import yfinance as yf
import plotly.graph_objects as go

indo = yf.download("INDO", start = "2023-01-01", end = "2024-09-30")
indo = indo.reset_index()

fig = go.Figure(data=[go.Candlestick(x=indo['Date'],
                open=indo['Open'],      #시가
                high=indo['High'],      #고가
                low=indo['Low'],        #저가
                close=indo['Close'])]   #종가
                )

# 설정을 건드렸을 때 update_layout을 적용하는지 update_traces를 적용하는지?
# 차트 제목, 축 label 설정

fig.update_layout(
    title = "INDO graph",
    xaxis_title = "Date",
    yaxis_title = "Price",
    hovermode = 'x',
    xaxis_rangeslider_visible = True
)

# 차트 색상 변경

fig.update_traces(
    increasing_line_color = 'black',
    decreasing_line_color = 'yellow'
)

fig.show()
```
