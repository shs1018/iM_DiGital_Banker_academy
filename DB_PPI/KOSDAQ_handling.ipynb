```python
import pandas as pd
```

```python
# 부르고 합치기
i = 0
while i <=3:
    tmp = pd.read_csv(f"data\data_202{i}.csv", usecols=['종목코드', '종목명', '소속부', '시가총액', '상장주식수'], encoding='cp949')
    tmp.rename(columns={'소속부': f'소속부_{i}', '시가총액': f'시가총액_{i}', '상장주식수': f'상장주식수_{i}'}, inplace = True)

    if i == 0:
        companies = tmp
    else:
        companies = pd.merge(companies, tmp, on=['종목코드', '종목명'], how='outer')
    i += 1
```

```python
companies['소속부_0'].value_counts()
```

```python
# 허용할 값 목록
allowed_values = [None, '중견기업부', '벤처기업부', '기술성장기업부', '관리종목(소속부없음)', '투자주의환기종목(소속부없음)']

# 각 소속부 열에서 허용된 값만 남기고 나머지 제거
for col in ['소속부_0', '소속부_1', '소속부_2', '소속부_3']:
    companies = companies[companies[col].isin(allowed_values) | companies[col].isnull()]
```
