# 데이터 무결성 확인
- 왜 같은 이름을 가진 기업들이 계속 보임 -> 아마 상호 변경으로 인한 것으로 추측됨

```python
tmp[tmp['회사명'] == '인바이츠바이오코아']
```

```python
tmp_closed = tmp[tmp['코넥스폐지일자'].notnull()] # 코넥스에서 폐지당한 애들 -> train
tmp_suv =  tmp[tmp['코넥스폐지일자'].isnull()] # 코넥스에서 폐지 안 당한 애들 -> target

print(tmp_closed.shape, tmp_suv.shape)
```

```python
def filter_numeric_columns(df):
    # 숫자형 칼럼만 필터링
    numeric_cols = df.select_dtypes(include=['number']).columns
    return df[numeric_cols]

# tmp 데이터프레임의 상관계수 계산
correlation_matrix_suv = filter_numeric_columns(tmp_suv).corr()
correlation_matrix_close = filter_numeric_columns(tmp_closed).corr()

# 상관관계가 높은 칼럼만 선택 (예: 절대값이 0.5 이상인 칼럼)
threshold = 0.5
high_corr_suv = correlation_matrix_suv[correlation_matrix_suv.abs() > threshold].dropna(axis=0, how='all').dropna(axis=1, how='all')
high_corr_close = correlation_matrix_close[correlation_matrix_close.abs() > threshold].dropna(axis=0, how='all').dropna(axis=1, how='all')

# 히트맵 그리기
plt.figure(figsize=(12, 10))  # 그림 크기 설정
plt.subplot(2, 1, 1)  # 첫 번째 서브플롯
sns.heatmap(high_corr_suv, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True)
plt.title('High Corr Matrix_Suv')

plt.subplot(2, 1, 2)  # 두 번째 서브플롯
sns.heatmap(high_corr_close, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True)
plt.title('High Corr Matrix_Closed')

plt.tight_layout()
plt.show()
```

```python
def filter_numeric_columns(df):
    # 숫자형 칼럼만 필터링
    numeric_cols = df.select_dtypes(include=['number']).columns
    return df[numeric_cols]

# tmp 데이터프레임의 상관계수 계산
correlation_matrix_suv = filter_numeric_columns(tmp_suv).corr()
correlation_matrix_close = filter_numeric_columns(tmp_closed).corr()

# 히트맵 그리기
plt.subplot(2,2,1)  # 그림 크기 설정
sns.heatmap(correlation_matrix_suv, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True)
plt.title('Corr Matrix_Suv')

plt.subplot(2,2,2)
sns.heatmap(correlation_matrix_close, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True)
plt.title('Corr Matrix_Closed')

plt.tight_layout()
plt.show()
```
