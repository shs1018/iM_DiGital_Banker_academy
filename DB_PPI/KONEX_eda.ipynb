```python
from pykrx import stock
import datetime

# 조회 날짜 설정
date = "20231231"

# KONEX 상장 종목 리스트 가져오기
tickers = stock.get_market_ticker_list(date=date, market="KONEX")

# 각 종목의 시가총액 정보 가져오기
data = []
for ticker in tickers:
    ohlcv = stock.get_market_ohlcv_by_date(date, date, ticker)
    if not ohlcv.empty:
        market_cap = ohlcv.iloc[0]['시가총액']
        data.append({
            '티커': ticker,
            '종목명': stock.get_market_ticker_name(ticker),
            '시가총액': market_cap
        })

# 결과 출력
for item in data:
    print(f"티커: {item['티커']}, 종목명: {item['종목명']}, 시가총액: {item['시가총액']}")
```

```python
import yfinance as yf
import pandas as pd

# KOSDAQ 종목 리스트 (예제: 일부 종목, 전체 종목은 직접 확보 필요)
kosdaq_tickers = [
    "035420.KQ",  # NAVER (예제)
    "041510.KQ",  # 에스엠 (SM Entertainment)
    "091990.KQ",  # 셀트리온헬스케어
    "066970.KQ"   # 엘앤에프 (L&F)
]  

# 2023-12-29 기준 데이터 조회 (31일 휴장 가능)
date = "2023-12-28"

# 데이터 저장 리스트
data = []

# 종목별 시가총액 조회
for ticker in kosdaq_tickers:
    stock_data = yf.Ticker(ticker)
    market_cap = stock_data.info.get("marketCap", "N/A")  # 시가총액 정보 가져오기

    data.append({"티커": ticker, "시가총액": market_cap})

# 데이터프레임 변환
df = pd.DataFrame(data)

# 시가총액 기준 정렬
#df = df.sort_values(by="시가총액", ascending=False)

# 결과 출력
print(df)

```
